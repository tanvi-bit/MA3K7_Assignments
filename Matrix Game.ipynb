```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as LA
import random
%matplotlib

rng = np.random.default_rng()
dim = 5
sample = 1e5
Det = []
for i in np.arange(sample):
    A=rng.integers(0,2, (dim,dim))
    detA = LA.det(A)
    Det.append(detA)

Max = max(Det) 
Min = min(Det) 
bins = np.arange(Min, Max+2)-0.5
fig, ax = plt.subplots(1,1)
Data, _, _=ax.hist(Det, bins, color ='pink', density=True, rwidth=0.8, edgecolor='black')
ax.set_xlabel('Det Game Matrix')
ax.set_ylabel('Probability distribution')
ax.set_xlim(Min-1, Max+1)
ax.set_xticks(np.arange(Min,Max+1))
plt.grid('on')
plt.show()
```

```python
rng = np.random.default_rng()

dim = 2

G = np.zeros((dim, dim))
for i in range(dim):
    for j in range(dim):
        G[i][j] = 2

print(G)
print("the number 2 is just a place holder for your values")

def check_human(x, G):
    while True:
        if G[x[0]][x[1]] == 2:
            return x
        else:
            print("Nope, choose again")
            i = (input('Please choose where to input your value in the form and separate the numbers of the index by a space '))
            x = i.split()
            for i in range(len(x)):
                # convert each item to int type
                x[i] = int(x[i])

def check_bot(x, G):
    while True:
        if G[x[0]][x[1]] == 2:
            return x
        else:
            x = [rng.integers(0, dim), rng.integers(0, dim)]
    
def game(myturn):
    global G
    global x
    if myturn == 1:
        i = (input('Please choose where to input your value in the form and separate the numbers of the index by a space: '))
        x = i.split()
        for i in range(len(x)):
            # convert each item to int type
            x[i] = int(x[i])
        x = check_human(x, G)
        G[x[0]][x[1]] = 0
        print(G)
        res = any(2 in sub for sub in G)
        if res:
            game(0)
        else:  
            detG = LA.det(G)
            if detG == 0:
                print("You win!")
            else:
                print("You lose :(")
            return 0
    
    else:
        x = [rng.integers(0, dim), rng.integers(0, dim)]
        x = check_bot(x, G)
        G[x[0]][x[1]] = 1
        print(G)
        res = any(2 in sub for sub in G)
        if res:
            game(1)
        else:  
            detG = LA.det(G)
            if detG == 0:
                print("You win!")
            else:
                print("You lose :(")
            return 0
myturn = 0 # 1 = player starts first, 0 = bot starts first

game(myturn)
```
